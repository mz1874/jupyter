```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-np.pi, np.pi, 1000)
y = np.cos(x)

plt.plot(x,y)
plt.show()
```

```python
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
y = np.tan(x)

plt.plot(x,y)
plt.show()
```

```python
y = np.sin(x)
"""
颜色支持 red, 或者十六进制
linestyle or ls
{'-', '--', '-.', ':', '', (offset, on-off-seq), ...}

"""
plt.plot(x,y,linestyle='--', linewidth=2, color="#451269",alpha = 0.5)
plt.show()
```

```python
y = np.sin(x)

# 可以设置仅仅控制某一个区间
# 设置坐标的范围
plt.xlim(0,3)
plt.ylim(-1,1)


"""
颜色支持 red, 或者十六进制
linestyle or ls
{'-', '--', '-.', ':', '', (offset, on-off-seq), ...}

"""
plt.plot(x,y,linestyle='--', linewidth=2, color="#451269",alpha = 0.5, )
plt.show()
```

```python
# 同样也可以显示X或者Y的刻度
#latex
x_val = [-np.pi, -np.pi/2, 0, np.pi/2 , np.pi]
v_text = ["π",r'$-\frac{\pi}{2}$', "0", "π/2", "π"]
#v_text = ["π", "-π/2", "0", "π/2", "π"]
plt.xticks(x_val,v_text)
plt.plot(x,y,linestyle='--', linewidth=2, color="#451269",alpha = 0.5, )
plt.show()
```

```python
# 设置坐标轴
x_val = [-np.pi, -np.pi/2, 0, np.pi/2 , np.pi]
v_text = ["π",r'$-\frac{\pi}{2}$', "0", "π/2", "π"]
#v_text = ["π", "-π/2", "0", "π/2", "π"]
plt.xticks(x_val,v_text)
# 修改y轴显示的数
plt.yticks([-1,-0.5, 0, 0.5,1])

# 获取全部坐标系
ax = plt.gca()
"""
支持的为 Top bottom left right
"""
ax.spines['top'].set_color("none")
ax.spines['right'].set_color("none")
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
plt.plot(x,y,linestyle='--', linewidth=2, color="#451269",alpha = 0.5, )
plt.show()

```
