```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-np.pi, np.pi, 1000)
y = np.cos(x)

plt.plot(x,y)
plt.show()
```

```python
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
y = np.tan(x)

plt.plot(x,y)
plt.show()
```

```python
y = np.sin(x)
"""
 red, 
linestyle or ls
{'-', '--', '-.', ':', '', (offset, on-off-seq), ...}

"""
plt.plot(x,y,linestyle='--', linewidth=2, color="#451269",alpha = 0.5)
plt.show()
```

```python
y = np.sin(x)

# 
# 
plt.xlim(0,3)
plt.ylim(-1,1)


"""
 red, 
linestyle or ls
{'-', '--', '-.', ':', '', (offset, on-off-seq), ...}

"""
plt.plot(x,y,linestyle='--', linewidth=2, color="#451269",alpha = 0.5, )
plt.show()
```

```python
# XY
#latex
x_val = [-np.pi, -np.pi/2, 0, np.pi/2 , np.pi]
v_text = ["",r'$-\frac{\pi}{2}$', "0", "/2", ""]
#v_text = ["", "-/2", "0", "/2", ""]
plt.xticks(x_val,v_text)
plt.plot(x,y,linestyle='--', linewidth=2, color="#451269",alpha = 0.5, )
plt.show()
```
