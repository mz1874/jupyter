```python
import matplotlib.pyplot as mp
import numpy as np

x = np.linspace(-np.pi, np.pi, 1000)
y = np.cos(x)

mp.plot(x,y, label = 'y = cos(x)')
mp.legend(loc = "center")

mp.show()
```

```python
# 
"""
marker 

facecolor 
s 
zorder 
"""
mp.scatter(0,1, marker='o',edgecolors ='red', facecolor="red", s=70, zorder= 3)
mp.plot(x,y, label = 'y = cos(x)')
mp.legend(loc = "center")

mp.show()
```
